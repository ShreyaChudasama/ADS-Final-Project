TextBlob is a Python (2 and 3) library for processing textual data. It provides a simple API for diving into common natural language processing (NLP) tasks such as part-of-speech tagging, noun phrase extraction and Classification.

```python
import nltk
nltk.download('brown')
nltk.download('punkt')
from textblob import TextBlob
```

```python
train = test2[['jobpost','Title']]
```

```python
train
```

```python
def converttext(bag):
    blob = TextBlob(bag)
    blob.noun_phrases
    text = ' '.join(blob.noun_phrases)
    return text
```

```python
train['jobpost'] = train['jobpost'].apply(lambda x: converttext(x))
```

```python
train['jobpost'] 
```

```python
train.values
```

#### Naive Bayes Classifier
Naive Bayes Classifier works on common principle, that every feature being classified is independent of the value of any other feature. 

```python
from textblob.classifiers import NaiveBayesClassifier
```

```python
cl = NaiveBayesClassifier(train.values)
```

#### Pickling of the file

```python
import pickle

object = cl
file = open('f2.obj','wb') #tried to use 'a' in place of 'w' ie. append
pickle.dump(object,file)
```

```python
import PyPDF2    #PyPDF2 is a Pure-Python Library built as a PDF toolkit.
pdfFileObj = open('dataset/resume_shruti.pdf','rb') 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
pdfReader.numPages
pageObj = pdfReader.getPage(0)  
text = pageObj.extractText()    # Extracted text from pdf file
#-----------------------------------------------
blob = TextBlob(text)
blob.noun_phrases
text = ' '.join(blob.noun_phrases)
text
#---------------------------------------------------
print(cl.classify(text))
prob_dist = cl.prob_classify(text)
rank = []
for k in profiles:
    ar = [k,prob_dist.prob(k)]
    rank.append(ar)
rank = sorted(rank, key=lambda x: x[1],reverse=True)
rank
```
